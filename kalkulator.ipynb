```python
import math

def penjumlahan():
  a= (float(input("Masukan nilai pertama: ")));
  b= (float(input("Masukan nilai kedua: ")));
  c= a+b;
  print(c);

def pengurangan():
  a= (float(input("Masukan nilai pertama: ")));
  b= (float(input("Masukan nilai kedua: ")));
  c= a-b;
  print(c);

def pembagian():
  a= (float(input("Masukan nilai pertama: ")));
  b= (float(input("Masukan nilai kedua: ")));
  c= a/b;
  print(c);

def perkalian():
  a= (float(input("Masukan nilai pertama: ")));
  b= (float(input("Masukan nilai kedua: ")));
  c= a*b;
  print(c);

def pangkat():
  a= (float(input("masukan bilangan: ")));
  b= (float(input("masukan pangkat: ")));
  c= math.pow(a, b);
  print(c);

def kuadrat():
  a= (float(input("masukan bilangan: ")));

  c= math.sqrt(a);
  print(c);

def logaritma():
  a= (float(input("masukan bilangan a (untuk a log b): ")));
  b= (float(input("masukan Bilangan b (untuk a log b): ")));

  c= math.log(b, a);

  print(c);
def sincostan():
  a= (int(input("masukan bilangan sin(derajat):  ")));
  d1= math.radians(a);
  c1= math.sin(d1);

  d2= math.radians(a);
  c2= math.cos(d2);

  d3= math.radians(a);
  c3= math.tan(d3);

  print("jadi sin= ",(c1));
  print("jadi cos= ",(c2));
  print("jadi tan= ",(c3));
def degree():
  b= (float(input("masukan bilangan(hasil berupa derajat) : ")));
  c= math.asin(b); #radian
  d= math.degrees(c); #derajat
  c2= math.acos(b); #radian
  d2= math.degrees(c2); #derajat
  print("derajat sin",(d))
  print("derajat cos",(d2))
  return 0;
def main():
  print("program kalkulator");
  print("1. penjumlahan");
  print("2. pengurangan");
  print("3. perkalian");
  print("4. pembagian");
  print("5. pangkat");
  print("6. kuadrat");
  print("7. logaritma");
  print("8. sincostan");
  print("9. Sin to degree");
  pilihan = (int(input("masukan pilihan anda: ")))
  if pilihan == 1:
    penjumlahan();
  elif pilihan == 2:
    pengurangan();
  elif pilihan == 3:
    perkalian();
  elif pilihan == 4:
    pembagian();
  elif pilihan == 5:
    pangkat();
  elif pilihan == 6:
    kuadrat();
  elif pilihan == 7:
    logaritma();
  elif pilihan == 8:
    sincostan();
  elif pilihan == 9:
    degree();
  else :
    while True:
      main();
  return pilihan;

if __name__ == "__main__":
    main()
```
